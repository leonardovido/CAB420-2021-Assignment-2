```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from sklearn.model_selection import train_test_split
```

## Load Data

```python
# Path to your 'en' folder with dataset.csv file
path = '/home/leonardo/Downloads/en/'

dataset = pd.read_csv (path + 'dataset.csv')
```

```python
print('Accents on the data base:\n')
print(dataset['accent'].value_counts())

fig = plt.figure(figsize=[12, 6])
ax = fig.add_subplot(1, 1, 1)
dataset['accent'].value_counts().plot(kind='bar')
ax.set_title('Dataset');
```

```python
# Ramdon sata split: 60% Training - 20% Testing - 20% Validation
train, test = train_test_split(dataset, test_size=int(.2*len(dataset)), random_state=1)

train, validation = train_test_split(train, test_size=int(.2*len(dataset)), random_state=1)
```

```python
# Check that we have samples for all the accents
print('\nAccents on the training set:\n')
print(train['accent'].value_counts())
print('\nAccents on the validation set:\n')
print(validation['accent'].value_counts())
print('\nAccents on the testing set:\n')
print(test['accent'].value_counts())
```

## Save data into .CSV

```python
train.to_csv('Data/training.csv', index=False)
validation.to_csv('Data/validation.csv', index=False)
test.to_csv('Data/testing.csv', index=False)
```
